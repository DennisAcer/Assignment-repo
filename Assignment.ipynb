```python

import numpy as np
import matplotlib.pyplot as plt

class Potential:
    def __init__(self, a, b, V0):
        self.a = a
        self.b = b
        self.V0 = V0

    def __call__(self, x):
        if x < 0 or x > self.b:
            return np.inf
        elif x <= self.a:
            return 0.0
        else:
            return self.V0

class Schrodinger:
    def __init__(self, potential, N):
        self.potential = potential
        self.N = N

    def solve(self, a, b):
        dx = b / (self.N + 1)
        x = np.linspace(dx, b - dx, self.N)
        H = np.zeros((self.N, self.N))
        for i in range(self.N):
            H[i, i] = 2 / dx**2 + self.potential(x[i])
            if i > 0:
                H[i, i - 1] = -1 / dx**2
            if i < self.N - 1:
                H[i, i + 1] = -1 / dx**2
        E, psi = np.linalg.eigh(H)
        return E, psi, x

# Parameters
a = 0.1
b = 1.0
V0 = 10
N = 1000

# Initialize classes
potential = Potential(a, b, V0)
schrodinger = Schrodinger(potential, N)

# Solve Schrödinger equation
E, psi, x = schrodinger.solve(a, b)

# Plot energy eigenvalues
plt.plot(np.arange(1, len(E) + 1), E, 'o')
plt.xlabel('Eigenvalue index')
plt.ylabel('Energy')
plt.title('Energy eigenvalues')
plt.show()

# Plot wavefunctions
for i in range(3):
    plt.plot(x, psi[:, i], label='n = {}'.format(i + 1))

plt.xlabel('Position')
plt.ylabel('Wavefunction')
plt.title('Wavefunctions')
plt.legend()
plt.show()
```
